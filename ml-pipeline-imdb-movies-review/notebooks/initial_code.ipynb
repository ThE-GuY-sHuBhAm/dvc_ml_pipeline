### Text preprocessing

```python
#lower-casing
df['review'] = df['review'].str.lower()
```

```python
#removing html tags
import re
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
# removing urls
def remove_urls(text):
    url_pattern = re.compile(r'http\S+|www\S+')
    return url_pattern.sub(r'', text)
```

```python
df['review'] = df['review'].apply(remove_urls)
```

```python
#removing punctuation
import string
exclude = string.punctuation
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, '')
        return text

df['review'] = df['review'].apply(remove_punctuation)
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

df['review'] = df['review'].apply(chat_conversion)
```

```python
#Remove Stopwords

import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')

def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
#Removing Emojis

import emoji
def demojize_safe(text):
    if pd.isna(text):
        return text
    return emoji.demojize(text)

df['review'] = df['review'].apply(demojize_safe)
```

```python
#Lemmatization
import nltk
nltk.download('wordnet')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
from nltk.stem import WordNetLemmatizer

def spacy_lemmatize_tokens(token_lists):
    lemmatized = []

    for tokens in token_lists:
        if isinstance(tokens, list):
            doc = nlp(" ".join(tokens))
            lemmatized.append([
                token.lemma_.lower()
                for token in doc
                if not token.is_stop and token.is_alpha
            ])
        else:
            lemmatized.append(tokens)

    return lemmatized

df["lemmas"] = spacy_lemmatize_tokens(df["review"].str.split())
df.head()
```

```python
df = pd.read_csv('tmdb_movies_preprocessed.csv')
```

```python
df['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import GradientBoostingClassifier

X_train = train_data.iloc[:, 0].values   # text column
y_train = train_data.iloc[:, -1].values  # label column

vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words="english"
)

X_train_vec = vectorizer.fit_transform(X_train)

clf = GradientBoostingClassifier(
    n_estimators=50,
    learning_rate=0.2
)

clf.fit(X_train_vec.toarray(), y_train)  # GB needs dense input
```

```python
y_preds = clf.predict(vectorizer.transform(test_data.iloc[:, 0].values).toarray())
```

```python
from sklearn.metrics import accuracy_score, r2_score
acc_score = accuracy_score(test_data.iloc[:, -1].values, y_preds)
```
